```python
companies = []
for i in range(1,501):
    url = "https://www.ambitionbox.com/list-of-companies?page={}".format(i)
    header = { "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) Chrome/120.0.0.0 Safari/537.36"}
    webPage = requests.get(url, headers=header).text

    soup = BeautifulSoup(webPage, 'html.parser')
    currPageCompany = soup.find_all('div',class_='companyCardWrapper')
    companies.extend(currPageCompany)
len(companies)
```

```python
name = []
rating = []
profile = []
address = []
reviews = []
salaries = []
interviews = []
jobs = []
benefits = []
photos = []

for company in companies:
    try:
        name.append(company.find('h2').text.strip())
    except:
        name.append(np.nan)

    try:
        rating.append(float(company.find(class_='rating_star_container').text.strip()))
    except:
        rating.append(np.nan)

    add = ''
    try:
        add = company.find(class_ = 'companyCardWrapper__interLinking').text.strip().split('|')
    except:
        add = [np.nan,np.nan]

    add = [a.strip() for a in add]
    try:
        profile.append(add[0])
    except:
        profile.append(np.nan)

    try:
        address.append(add[1])
    except:
        address.append(np.nan)

    try:
        reviews.append(company.find_all(class_='companyCardWrapper__ActionWrapper')[0].find('span', class_='companyCardWrapper__ActionCount').text.strip())
    except:
        reviews.append(np.nan)

    try:
        salaries.append(company.find_all(class_='companyCardWrapper__ActionWrapper')[1].find('span', class_='companyCardWrapper__ActionCount').text.strip())
    except:
        salaries.append(np.nan)

    try:
        interviews.append(company.find_all(class_='companyCardWrapper__ActionWrapper')[2].find('span', class_='companyCardWrapper__ActionCount').text.strip())
    except:
        interviews.append(np.nan)

    try:
        jobs.append(company.find_all(class_='companyCardWrapper__ActionWrapper')[3].find('span', class_='companyCardWrapper__ActionCount').text.strip())
    except:
        jobs.append(np.nan)

    try:
        benefits.append(company.find_all(class_='companyCardWrapper__ActionWrapper')[4].find('span', class_='companyCardWrapper__ActionCount').text.strip())
    except:
        benefits.append(np.nan)

    try:
        photos.append(company.find_all(class_='companyCardWrapper__ActionWrapper')[5].find('span', class_='companyCardWrapper__ActionCount').text.strip())
    except:
        photos.append(np.nan)
```

```python
data = {'Name' : name, 
      'Rating' : rating,
      'Profile' : profile,
      'Address' : address,
      'Reviews' : reviews,
      'Salary' : salaries,
      'Jobs' : jobs,
      'Benefits' : benefits,
      'Photos' : photos
      }
final = pd.DataFrame(data)
final
```

```python
final.to_csv('companyData.csv',index=False)
```
