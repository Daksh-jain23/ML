# Why NumPy ?

### Speed  
NumPy is written in C and uses vectorized operations, performing tasks on entire arrays at once. This makes it 10–100× faster than pure Python loops.

###  Memory Efficiency  
NumPy arrays store fixed-type data in a compact, continuous memory block, while Python lists store references, using more memory.

### Simplicity  
Operations in NumPy are applied directly on arrays without explicit loops, making code shorter, cleaner, and easier to read.

### Mathematical Power  
NumPy provides built-in functions for statistics, linear algebra, and scientific computing — all highly optimized in C/C++.

### Multi-dimensional Support 
NumPy supports multi-dimensional arrays (matrices, tensors), which pure Python handles inefficiently with nested lists.


| Feature | Pure Python | NumPy |
|----------|--------------|--------|
| Speed | Slow | Fast |
| Memory | High usage | Efficient |
| Syntax | Verbose | Concise |
| Multi-Dimensional Arrays | Difficult | Built-in |
| Math Functions | Limited | Extensive |

### Operation

Slicing : np[ start : end : step ]

```python
a = np.arange(10)
print(f"a         = {a}")

#access 5 consecutive elements (start:stop:step)
c = a[2:7:1];     print("a[2:7:1] = ", c)

# access 3 elements separated by two 
c = a[2:7:2];     print("a[2:7:2] = ", c)

# access all elements index 3 and above
c = a[3:];        print("a[3:]    = ", c)

# access all elements below index 3
c = a[:3];        print("a[:3]    = ", c)

# access all elements
c = a[:];         print("a[:]     = ", c)
```
