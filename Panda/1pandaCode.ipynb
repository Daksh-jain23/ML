```python
import pandas as pd

print(pd.__version__)
```

```python

#Series = 1D labeled array 
data = [151,102,103]
series = pd.Series(data)
print(series)
print()

data2 = [1.2]
series2 = pd.Series(data2)
print(series2)
print()
```

```python
data = [151,102,103]
series_labeled = pd.Series(data, index=['room no','price', 'floor']) #length of index == length of data

print(series_labeled)

print(series_labeled.loc['price'])  #if not exists there will be an error

series_labeled.loc['price'] = 200
print(series_labeled.loc['price'])

print(series_labeled.iloc[1])        #access via index


```

```python
data = [51,102,201,55,36]
series = pd.Series(data, index=['a','b','c','d','e'])
print(series[series > 100])
print()

```

```python
# Data Frames = 2d tabular array (labeled column)

data = {"Pokemon" : ["Balbausaur","Squirtle","Venasuar","Charizard"],
        "Attack" : [20, 10, 30, 90],
        "Trainer": ["Ash", "Misty", "Rock", "Gary"]
        }

df = pd.DataFrame(data)
print(df)
```

```python
df["Win"] = pd.Series([1,2,3,5])
print(df)
```

```python
df = pd.concat([df,  pd.DataFrame([{"Pokemon" : "Blastoise",\
                                    "Attack": 77,\
                                    "Trainer":"Misty",
                                    "Win" : 10}])])

print(df) 
```
